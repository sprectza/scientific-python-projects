```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 10000

sx = np.zeros(N)
sy = np.zeros(N)

for i in range (1, N):
    # Creates a random integer between 1 and 4
    k = np.random.randint(1,4)
    
    # update x and y 
    sx[i] = sx[i-1] / 2 + k - 1
    sy[i] = sy[i-1] / 2
    
    if k == 2:
        sy[i] += 2
        
plt.plot(sx, sy, 'k.', markersize=1) # k. means black dots, and marker size tells the size of dots 
plt.title('Sierpinski Triangle')
plt.axis('off')
plt.show()
```
